```python
file=open('week1_edges.txt','r')
c=file.read()
T = c.split(sep='\n')
T = [t.split(sep='\t') for t in T]
T = [t[0].split(sep=' ') for t in T[:-1]]
T = [ [int(t[0]),int(t[1]), int(t[2])] for t in T[1:]]
```

```python
def findE():
    global T
    global V
    w=float('inf')
    for i in range(len(T)):
        if(T[i][0] in V and T[i][1] not in V and T[i][2]<w):
            v = T[i][1]
            edg = T[i]
            w = T[i][2]
        elif(T[i][0] not in V and T[i][1] in V and T[i][2]<w):
            v = T[i][0]
            edg = T[i]
            w = T[i][2]
    return v, edg, w
        
```

```python
#  = [[1,2,1],[2,4,2],[3,1,4],[4,3,5],[4,1,3]]
nV=500
nE=2184
V=[1]
E=[]
score=0

while len(V)<nV:
    v,edg,w = findE()
    V.append(v)
    E.append(edg)
    score+=w

# print('V are ', V)
# print('Edges are ', E)
print('Score is ', score)


```
