```python
import math
import random
```

```python
file=open('week4_test1.txt','r')
c=file.read()
T = c.split(sep='\n')
nV=int(T[0])
T = [t.split(sep=' ') for t in T[1:]]
T = [ [int(t[0]),int(t[1])]  for t in T]
```

```python
T
```

```python
def checkClauses(r):
    global T
    C=[]
    flag=True
    for t in T:
        r1=r[t[0]-1] if t[0]>0 else (not r[-t[0]-1])
        r2=r[t[1]-1] if t[1]>0 else (not r[-t[1]-1])
        print(r1,r2)
        if r1 is not r2:
            C.append(t)
            flag=False
    return flag,C
        

            
```

```python
def main():
    global nV
    for i in range(round(math.log2(nV))):
        r=random.choices([0,1],[1,1], k=nV)
        for j in range(2*nV**2):
            flag,wrongClauses=checkClauses(r)
            if flag:
                return True
            else:
                c=random.choice(wrongClauses)
                v=random.choice([0,1])
                idx=c[v]
                r[abs(idx)-1]=not r[abs(idx)-1]
                

    return False
```

```python
result=main()
print(result)
```
