Lecture Notes:

1. General
Y = f(X)+epsi
- Reducible and irreducible errors
- MSE
- Bias and variance balance
- Sample size: n
- Predictor dim: p


2. Linear regression
Y = b0+b1*X+epsi
- Coefficients
- TSS (Total sum of square): depends on Y
- RSS (Residual sum of square): least square
- SE (Standard Error): depends on X and epsi; for each coeff; compute by using RSE (or RSS)
- t-stats: depends on coeff and its SE; for each coeff
- p-value: for each coeff;large p indicates coeff=0
- Confidence interval: for each coeff
- RSE (Residual standard error): depends on RSS,n,p; measures the standard deviation of epsi
- R-square (0-1): depends on TSS and RSS; measures the propotion of variance explained

3. Multiple Linear regression
Y = b0+b1*X1+b2*X2+...(bp-1)*Xp+epsi
- F-stats: depends on TSS,RSS,n,p; for all coeff
- Variable selection: forward selection, backward selection, mixed selection
- Prediction interval
- Qualitative predictors (0 or 1)

4. Potential problems
- Non-linearity data: use residual plots to identify dicernible pattern
- Correlation of error terms: use residual plot as a function of time to idenfity dicernible pattern
- Non-constant variance of error terms: use residual plot to identify a funnel shape; transform Y using concave functions
- Outliers: use residual plot to eyeball them
- High leverage points: compute leverage stats and compare with (1+p)/n; depends on X and n
- Collinearity: look at the correlation matrix of the predictors; compute VIF (variance inflation factor); VIF=1 indicates no collinearity, VIF>5 or >10 indicates collinearity; drop one predictor or combine them into a new one.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
import findspark
from pyspark import SparkContext
from pyspark.sql import SparkSession, DataFrame, DataFrameReader
from pyspark.ml.regression import LinearRegression, LinearRegressionModel, LinearRegressionSummary
from pyspark.mllib.regression import LabeledPoint
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
from pyspark.ml.evaluation import RegressionEvaluator

%matplotlib inline

# change the path on your machine
findspark.init("/Users/lhd0430/Downloads/spark")
```

```python
# Creat spark session
spark = SparkSession.builder \
   .master("local") \
   .appName("Linear Regression Model") \
   .config("spark.executor.memory", "1gb") \
   .getOrCreate()
```

```python
# Load data as pyspark.sql.DataFrame
data = spark.read.csv("../data/Boston.csv", header=True, inferSchema=True)
data.cache()
print(data.count())
data = data.dropna()
print(data.count())
```

```python
# LR

# Convert feature to vector type
vecAssembler = VectorAssembler(inputCols=["lstat"], outputCol="features")
df = vecAssembler.transform(data)

# Fit model
lr = LinearRegression(featuresCol="features", labelCol="medv")
ml = lr.fit(df)
print(ml.intercept)
print(ml.coefficients)

# Predict
predict = ml.transform(df)

# Evaluate
evaMSE = RegressionEvaluator(labelCol="medv",metricName="mse")
evaR2 = RegressionEvaluator(labelCol="medv",metricName="r2")
mse = evaMSE.evaluate(predict)
r2 = evaR2.evaluate(predict)

print(r2)
print(mse)

# Plot
# Convert to pandas dataframe
pddf = predict.toPandas()

plt.figure()
plt.scatter(pddf.lstat,pddf.medv,c='b')
plt.scatter(pddf.lstat,pddf.prediction,c='r')
plt.show()
```

```python
# MLR

# Convert feature to vector type
vecAssembler = VectorAssembler(inputCols=["lstat","tax"], outputCol="features")
df = vecAssembler.transform(data)

# Correlation matrix
corr = Correlation.corr(df,'features')
print(corr.collect()[0][0])

# Fit model
lr = LinearRegression(featuresCol="features", labelCol="medv")
ml = lr.fit(df)
print(ml.intercept)
print(ml.coefficients)

# Predict
predict = ml.transform(df)

# Evaluate
evaMSE = RegressionEvaluator(labelCol="medv",metricName="mse")
evaR2 = RegressionEvaluator(labelCol="medv",metricName="r2")
mse = evaMSE.evaluate(predict)
r2 = evaR2.evaluate(predict)

print(r2)
print(mse)

# Plot
# Convert to pandas dataframe
pddf = predict.toPandas()

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter3D(pddf.lstat,pddf.tax,pddf.medv,c='b')
ax.scatter3D(pddf.lstat,pddf.tax,pddf.prediction,c='r')
plt.show()
```
